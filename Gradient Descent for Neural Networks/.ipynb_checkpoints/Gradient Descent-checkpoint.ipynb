Note: In a regression function, y=wx+b; w ==> Weight and b ==> Bias

***Training Process:***<br>
1. Firstly, you initialize some random weights and bias to the regression function and then then calculate the score using a neuron by giving it some input values. This is called a Forward Pass.
2. Now that you have got your predicted value, you find the error by using a loss function. For logistic problems, we generally use Binary Cross Entropy loss function. Similarly, you find the loss of all the input values for the same weight and bias and then calculate its mean. You get your cost function. After the first epoch (by the way, one epoch is going through all the training samples once), you get your loss and your goal is now to back propagate such that you adjust your weights and bias so that the total loss is less than the previous loss. Hence, to get a better model, you need to have multiple epochs.
3. Now, we need to change the weights and bias such that there is reduction in the total loss. To adjust them, you can subtract them with the slope of the gradient descent (It is the graph between the loss function(y) and the parameters like weights(x1,x2,..xk) and bias(xk+1)) multiplied with some value called the learning rate.<br><img src='eq.jpg' width=500><br>Here the derivative can also be read as the change in loss per change in the weights or the bias. Basically, learning rate is used to adjust the derivative value as the derivative may be large and all that we wanted is a small change. Generally, the value of learning rate is 0.01 . We use derivative instead of a small constant value because slope varies with the value of 'x' and hence, when the 'x' is large, we get large value to be reduced and hence, there is large change in the 'weights/bias' and when the value of 'x' gets near the global minima, the slope becomes very small and hence, we require small differences between the previous and the current weights/bias and hence we are now taking small steps to reach the global minima. <br><img src='gradient_descent.jpg' width=500><br>
4. Now that you have back propagated after the first epoch, we adjust our weights and the bias based on the derivative and learning rate equation and then again do a forward pass and calculate the score. Then you calculate the error using the loss function and take a mean of it. This is the end of the second epoch. Right now, we are using Batch Gradient, other gradients like the stochastic gradient and the mini-batch gradient uses slightly different technique. In batch graident descent you take the entire training samples and calculate the error jsut like the current training process.
5. The entire process is then repeated for the mentioned number of epochs.

```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
df=pd.read_csv('insurance_data.csv')
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[['age','affordibility']],df['bought_insurance'],test_size=0.2)
```

```python
x_train['age']/=100
x_test['age']/=100
```

```python
model= keras.Sequential([
#   output_size=1 because we need to know whether the person buys insurance or not, kernel_initializer=1 because initially
#   we are going to set the weights to 1 and bias_initializer=0 because initially we are going to set the bias to 0.
    keras.layers.Dense(1,input_shape=(2,),activation='sigmoid',kernel_initializer='ones',bias_initializer='zeros')
])
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train,epochs=5000)
```

```python
model.evaluate(x_test, y_test)
```

```python
def sigmoid(x):
    import math
    return 1/(1+math.exp(-x))
```

```python
def prediction_function(age,affordibility):
    weighted_sum=weights[0]*age+weights[1]*affordibility+bias[0]
    return sigmoid(weighted_sum)
```

```python
prediction_function(0.27,0)
```

```python
# Our prediction function is working great. Now lets make a function to implement gradient descent.
```

```python
import numpy as np
def binary_crossentropy(y_true,y_predicted):
    epsilon=1e-15
    y_predicted=[max(i,epsilon) for i in y_predicted]
    y_predicted=[min(i,1-epsilon) for i in y_predicted]
    return -np.mean(y_true*np.log(y_predicted)+(np.ones(len(y_predicted))-y_true)*np.log(np.ones(len(y_predicted))-y_predicted))
```

```python
# The previous sigmoid function was used to calculate individual sigmoid values. To do this in an array we are going to 
# create using numpy library.
def sigmoid_array(x):
    return 1/(1+np.exp(-x))
```

```python
def gradient_descent(age,affordibility,y_true,epochs):
    # Initializing the initial weights and bias
    w1=w2=1
    bias=0
    learning_rate=0.5
    
    for i in range(epochs):
        # Calculate the weighted sum
        weighted_sum=w1*age+w2*affordibility+bias
        # Calculate its prediction using the sigmoid as the activation function
        y_predicted=sigmoid_array(weighted_sum)
        # Finding the loss
        loss=binary_crossentropy(y_true,y_predicted)
        # Finding the derivatives
        w1_derivative=np.mean(age*(y_predicted-y_true))            # Derivative of the loss function
        print(f"w1_derivative:{w1_derivative}")
        w2_derivative=np.mean(affordibility*(y_predicted-y_true))
        print(f"w2_derivative:{w2_derivative}")
        bias_derivative=np.mean((y_predicted-y_true))
        print(f"bias_derivative:{bias_derivative}")
        # Adjusting the weights and the bias
        w1=w1-w1_derivative*learning_rate
        w2=w2-w2_derivative*learning_rate
        bias=bias-bias_derivative*learning_rate
        print(f"epoch: {i}   w1:{w1}  w2:{w2}  bias:{bias} loss:{loss}")
        print("")
    return w1, w2, bias
```

```python
gradient_descent(x_train['age'], x_train['affordibility'],y_train,epochs=1000)
```
