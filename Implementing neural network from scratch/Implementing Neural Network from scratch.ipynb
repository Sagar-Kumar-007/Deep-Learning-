```python
X_train['age']=X_train['age']/100
X_test['age']=X_test['age']/100
```

```python
import numpy as np
def sigmoid_array(x):
    return 1/(1+np.exp(-x))
```

```python
def binary_crossentropy(y_pred,y_true):
    epsilon=1e-15
    y_pred=[max(i,epsilon) for i in y_pred]
    y_pred=[min(i,i-epsilon) for i in y_pred]
    return -np.mean(y_true*np.log(y_pred)+(np.ones(len(y_true))-y_true)*np.log(np.ones(len(y_true))-y_pred))
```

```python
def gradient_descent(age,affordibility,y_true,epochs):
    # Initializing weights and bias:
    w1=w2=1
    bias=0
    learning_rate=0.01
    for i in range(epochs):
        weighted_sum=w1*age+w2*affordibility+bias
        y_pred=sigmoid_array(weighted_sum)
        loss=binary_crossentropy(y_pred,y_true)
        w1_derivative=np.mean(age*(y_pred-y_true))
        w2_derivative=np.mean(affordibility*(y_pred-y_true))
        bias_derivative=np.mean(y_pred-y_true)
        w1=w1-learning_rate*w1_derivative
        w2=w2-learning_rate*w2_derivative
        bias=bias-learning_rate*bias_derivative
        print(f"epoch: {i} loss:{loss}")
    return w1, w2, bias
```

```python
gradient_descent(X_train['age'],X_train['affordibility'],y_train,5000)
```

```python
# Lets create our own neural network:
```

```python
class myNN:               # Defining class
    def __init__(self):   # This is a constructor where variables can be defined. self.something is used when you want to use
                          # it as the instance for a class
        self.w1=1
        self.w2=1
        self.bias=0
    def gradient_descent(self,age,affordibility,y_true,epochs):  # While defining an object we use self
        # Initializing weights and bias:
        w1=w2=1
        bias=0
        learning_rate=0.5
        for i in range(epochs):
            weighted_sum=w1*age+w2*affordibility+bias
            y_pred=sigmoid_array(weighted_sum)
            loss=binary_crossentropy(y_pred,y_true)
            w1_derivative=np.mean(age*(y_pred-y_true))
            w2_derivative=np.mean(affordibility*(y_pred-y_true))
            bias_derivative=np.mean(y_pred-y_true)
            w1=w1-learning_rate*w1_derivative
            w2=w2-learning_rate*w2_derivative
            bias=bias-learning_rate*bias_derivative
            print(f"epoch: {i} loss:{loss}")
        return w1, w2, bias
    def fit(self,X,y,epochs):
        self.w1, self.w2, self.bias=gradient_descent(X['age'],X['affordibility'],y,epochs)
    def predict(self,X_test):
        weighted_sum=self.w1*X_test['age']+self.w2*X_test['affordibility']+self.bias
        return sigmoid_array(weighted_sum)
```

```python
model=myNN()
model.fit(X_train,y_train,10000)
```
