CNN first divides the image into different sets of features.

***How does the computer identify these tiny features?***<br>
<img src='a.jpg' width=700><br>
To identify the features, we use a concept of filters. In this case, we have three filters, the loop, the vertical line and the diagonal line. Now we take the matrix of each filter (we have taken the size to be 3*3 in this case) and then multiply it with the original image filter and take the mean. You can hence see that the part of the image containing the pattern would have the value to be one and the part closely related to that pattern would be near to one and the ones that are not related will be far away from one. The matrix thus formed after multiplication is called the feature map. Since we are multiplying each pattern matrix with every row and column our stride is one. If we multiplied our pattern matrix after every two rows, the value of stride would have been 2.Since we are recognizing the pattern from the feature map, it does not matter whether the image is moved to the left/ right or any other place, but it matters if the image is rotated or scaled. We will see later how to tackle rotated or scaled images in CNN.<br>
<img src='b.jpg' width=700><br>
Example for loopy pattern detection.<br>
<img src='d.jpg' width=700><br>
This is how the neural network is to be created. The first part is the feature extraction, where CNN layers are applied to detect the features and make a feature map. The second part is the classification, where you flatten the feature map and then you train the model just like any other ordinary ANN.
<img src='e.jpg' width=900><br>

While constructing the feature map, we use the **ReLu** activation function because it converts the values that are not related with the pattern to zero (because ReLu converts negative values to zero) and the values that are related to the pattern are left unchanged (because ReLu does not change the positive values to anything).<br><img src='f.jpg'><br>
***ReLu helps in making the model Non-Linear.***<br>
We have not discussed about lowering the computation part yet. We can still observe that the number of pixels remain the same because our input image and the feature map is still of the same size.

***To reduce the computation power, we will use a method called Pooling.***<br>
We introduce the pooling layers after each feature map to reduce the size. There are three types of pooling layers:<br>
1. Max Pooling: Max pooling is widely used because it detects the pixels that are most prominent.<br>
<img src='g.jpg' width=500><br>Stride 2 means we are shifting our matrix 2 places when comparing the max values.<br>
<img src='h.jpg' width=700> You can also see the same thing we are performing for the loopy pattern. When you use pooling layers after the feature map, you can see that the position of the image does not matter unless the image is scaled or rotated.<br>
2. Average Pooling: Similar to max pooling but you average all the values of the feature map that comes within the size of the matrix that is later shifted using the value of stride.
3. 

***Thus the pattern followed when performing CNN:***<br>
**A. Feature Extraction:<br>
&nbsp; 1. Create the filters and construct a feature map.<br>
&nbsp; 2. Now use the ReLu activation function to remove the irrelevant patterns from the feature map.<br>
&nbsp; 3. Now create the pooling layers to reduce the dimensions and hence the computation power.<br>
B. Classification:<br>
&nbsp; 1. Flatten the feature map in order to train the neural network model.<br>
&nbsp; 2. Now train the model using the ANN and back propagation.<br> Do not forget to convert the RGB values from 0-255 to 0-1 by dividing it by 255.**

<img src='i.jpg'>

***Model is tolerant to variations and distortions*** is applicable in the case when you use ***Max Pooling*** because it takes the prominent feature and out of the matrix that is shifted with the help of the stride. Hence, Max Pooling is widely used as it eliminates the variations in the image.

This is how a complete CNN looks like:<br><img src='j.jpg'><br>
There could be 'n' number of (Convolution layers+Relu) and pooling but for our convenience we are using only two layers of convolution layers and pooling

In the image, we are making the filters for eyes, nose and ears but in reality it takes all the pixels and make them into filters.

<br><br><img src='k.jpg'>

Connection Sparsity: It is unlike the ANN where you take all the pixels/ cells of the matrix image and use them as features. In CNN you use filters which strides through the image and take the prominent features using ReLu.<br>
Parameter Sharing: It means that once the pattern is trained, you can use that to recognize the pattern in any image.

<img src='l.jpg'><br>
Data Augmentation technique means that you must manually rotate or scale the images before training.

***While Training the CNN, it automatically detects the filters through Back Propagation. Hyper Parameter tuning is done to the size of each filter and the number of filters used.***

We use Categorical Cross Entropy if our dependent features are one-hot encoded. If our dependent variables and just a number we will use Sparse Categorical Cross Entropy.

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test)=tf.keras.datasets.cifar10.load_data()
```

```python
plt.imshow(x_train[0])
```

```python
y_train[0]
```

```python
y_train[:5]
```

```python
# You can see that our dependent features are in 2D. So first we need to convert it into 1D.
```

```python
y_train.shape
```

```python
y_train=y_train.reshape(-1,) # -1 means having the same shape for that dimension
y_train[:5]
```

```python
y_test=y_test.reshape(-1,)
y_test[:5]
```

```python
classes=['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']
```

```python
classes[y_train[0]]
```

```python
x_train.shape
```

```python
x_train[0]
```

```python
x_test.shape
```

```python
# Now we will perform feature engineering by dividing the RGB values by 255 so that the values now range between 0 and 1.
x_train=x_train/255
```

```python
x_test1=x_test/255
```

```python
# Now we will train our model.
```

```python
model=keras.Sequential([
    # Cnn layers+pooling:
    keras.layers.Conv2D(filters=32,kernel_size=(3,3),activation='relu',input_shape=(32,32,3)),
    keras.layers.MaxPooling2D(),
    
    keras.layers.Conv2D(filters=64,kernel_size=(3,3),activation='relu'),
    keras.layers.MaxPooling2D(),
    
    # ANN:
    keras.layers.Flatten(),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dropout(rate=0.5),
    keras.layers.Dense(10,activation='softmax') # softmax considers probability and hence the sum of all softmax values will be one for
])

model.compile(optimizer='adam', metrics=['accuracy'],loss='sparse_categorical_crossentropy')
model.fit(x_train, y_train,epochs=100)
```

```python
# You can see that our accuracy is very high even for small number of epochs.
```

```python
model.evaluate(x_train, y_train)
```

```python
model.evaluate(x_test1, y_test)
```

```python
y_pred=model.predict(x_test1)
```

```python
y_pred[0]
```

```python
# The max value of the 10 will be the answer
y_pred=[np.argmax(i) for i in y_pred]
```

```python
y_pred[:5]
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
